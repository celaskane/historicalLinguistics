# **2. Recontrução de Árvores Filogenéticas**

```python
# Definindo alguns idiomas
languages = ['Norwegian','Swedish','Icelandic','Dutch','English']
```

```python
# Importando a função squareform
from lingpy.algorithm import squareform
```

```python
# Definindo uma matriz de distâncias
distances = squareform([0.5,0.67,0.8,0.2,0.4,0.7,0.6,0.8,0.8,0.3])
```

```python
# Realizando uma análise de agrupamento (clustering) 
# método Neighbor Joining (NJ)

tree = neighbor(distances, languages)
print(tree)
```

```python
# Exibindo a árvore em ASCII-Art
tree = Tree(tree)
print(tree.asciiArt())
```
